# 폴더에 있는 json파일들이 얼마나 있는지 확인

```python
import os

# 검색할 폴더 경로 설정
target_folder = '/content/drive/MyDrive/Colab Notebooks/finalproject/data_interview_6400'

# JSON 파일 수를 세는 함수
def count_json_files(directory):
    json_count = 0
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                json_count += 1
    return json_count

# JSON 파일 수 계산
json_file_count = count_json_files(target_folder)

print(f"'{target_folder}' 폴더 내의 JSON 파일 수: {json_file_count}")
```

```python
import os
import json

folder_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/data_interview_1000'
error_files = []

intents_data = []

for file in os.listdir(folder_path):
    if file.endswith('.json'):
        file_path = os.path.join(folder_path, file)
        try:
            with open(file_path, 'r') as f:
                data = json.load(f)
                intents = data.get("dataSet", {}).get("answer", {}).get("intent", [])
                for intent in intents:
                    intent_info = {
                        "text": intent.get("text", "N/A"),
                        "expression": intent.get("expression", "N/A"),
                        "category": intent.get("category", "N/A")
                    }
                    intents_data.append(intent_info)
        except json.JSONDecodeError:
            error_files.append(file)

print(f"Processed {len(intents_data)} intents.")
if error_files:
    print("The following files had errors and were skipped:")
    for file in error_files:
        print(file)
```

```python
import csv
from openai import OpenAI

# OpenAI API 키 설정
client = OpenAI(api_key="")

# 원본 CSV 파일 읽기
input_csv_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_1000.csv'
output_csv_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_label_1000.csv'

# 결과를 저장할 새로운 CSV 파일 생성
with open(input_csv_path, mode='r', newline='', encoding='utf-8') as input_file, \
     open(output_csv_path, mode='w', newline='', encoding='utf-8') as output_file:
    csv_reader = csv.reader(input_file)
    csv_writer = csv.writer(output_file)
    # 헤더 추가
    csv_writer.writerow(["Question Text", "Answer Text", "Category"])

    next(csv_reader)  # 헤더 스킵
    for row in csv_reader:
        question_text, answer_text = row
        try:
            # GPT에게 질문 분류 요청
            response = client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=[
                    {
                        "role": "user",
                        "content": f"Classify the following question based on its context: If the Question is HR department perspective, it is output as '1', if it is practical questions in ICT, it is output as '2', and if it is personality evaluation based on self-introduction, it is output as '3'\
                        The output only produces numbers.\
                        \nQuestion: {question_text}"
                    }
                ],

                temperature=0.5,
                max_tokens=60,
                top_p=1.0,
                frequency_penalty=0.0,
                presence_penalty=0.0
            )
            # 응답에서 카테고리 추출
            category = response.choices[0].message.content
            print(f"Question: {question_text}, Category: {category}")
            # 결과 CSV 파일에 기록
            csv_writer.writerow([question_text, answer_text, category])
        except Exception as e:
            print(f"Error processing question: {question_text}. Error: {e}")
            # 오류 발생 시, 카테고리 없이 기록
            csv_writer.writerow([question_text, answer_text, "Error"])
```

```python
import csv
import random
from collections import defaultdict

# 원본 CSV 파일 경로
input_csv_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_label_1000.csv'

# 결과를 저장할 CSV 파일 경로
output_csv_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_sampled_100.csv'

# 카테고리별로 데이터를 분류하여 저장할 딕셔너리
data_by_category = defaultdict(list)

# 원본 CSV 파일 읽기 및 데이터 분류
with open(input_csv_path, mode='r', newline='', encoding='utf-8') as input_file:
    csv_reader = csv.reader(input_file)
    next(csv_reader)  # 헤더 스킵
    for row in csv_reader:
        if len(row) < 3:  # 카테고리 정보가 없는 행은 건너뜀
            continue
        category = row[2]  # 세 번째 열이 카테고리
        data_by_category[category].append(row)

# 각 카테고리별로 추출할 데이터 수 설정
sample_sizes = {
    '1': 30,  # 3의 비율
    '2': 50,  # 5의 비율
    '3': 20  # 2의 비율
}

# 결과 데이터 선택
selected_data = []
for category, samples in sample_sizes.items():
    if len(data_by_category[category]) >= samples:
        selected_data.extend(random.sample(data_by_category[category], samples))
    else:
        print(f"Warning: Not enough data in category '{category}' to sample {samples} items.")

# 선택된 데이터를 새로운 CSV 파일에 저장
with open(output_csv_path, mode='w', newline='', encoding='utf-8') as output_file:
    csv_writer = csv.writer(output_file)
    csv_writer.writerow(["Question Text", "Answer Text", "Category"])  # 헤더
    for row in selected_data:
        csv_writer.writerow(row)

print(f"Sampled data saved to {output_csv_path}")
```

```python
import csv
import os
from openai import OpenAI

# OpenAI API 키 설정
client = OpenAI(api_key="")

input_csv_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_sampled_100.csv'
output_csv_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_sampled_feedback_100.csv'

def generate_feedback(question, answer):
    response = client.chat.completions.create(
      model="gpt-3.5-turbo",
      temperature=0.7,
      messages=[
          {
                "role": "system",
                "content": "This system analyzes the interviewer's questions and the applicant's answers to provide applicants with advice to help them pass the interview.\
                             When providing advice to an applicant, start with 'Interview Feedback: 지원자님은' and express it in line text.\
                             Avoid using unprofessional or casual language and frame your advice using professional terminology appropriate to the interview situation.\
                             Please speak in Korean"
            },
          {
              "role": "user",
              "content": f"Please provide applicants with advice to help them pass the interview based on the following questions and answers. Please speak in Korean\
                          \n질문: {question}\n답변: {answer}"
          }
      ],
    )
    feedback_text = response.choices[0].message.content
    return feedback_text.replace('\n', ' ')

# 원본 CSV 파일 읽기 및 결과 파일에 쓰기
with open(input_csv_path, mode='r', newline='', encoding='utf-8') as input_file, \
     open(output_csv_path, mode='w', newline='', encoding='utf-8') as output_file:
    csv_reader = csv.reader(input_file)
    csv_writer = csv.writer(output_file)

    # 헤더 읽기 및 새 헤더 작성
    headers = next(csv_reader)
    csv_writer.writerow(headers + ["Feedback"])

    for row in csv_reader:
        question, answer = row[0], row[1]
        feedback = generate_feedback(question, answer)
        csv_writer.writerow(row + [feedback])
        print(f"Processed feedback for question: {question}")
        print(f"feedback: {feedback}")

print("Finished generating feedback for all questions.")
```

```python
import csv
import json

# CSV 파일 경로
csv_file_path = '/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_sampled_feedback_100.csv'

# CSV 파일을 읽고 각 row를 dictionary로 변환
results = []
with open(csv_file_path, mode='r', encoding='utf-8') as csv_file:
    csv_reader = csv.reader(csv_file)

    # 헤더(첫 번째 행) 건너뛰기 (필요한 경우)
    next(csv_reader)

    for row in csv_reader:
        new_entry = {
            "messages" : [
                {"role": "system", "content": "This system analyzes the interviewer's questions and the applicant's answers to provide applicants with advice to help them pass the interview. When providing advice to an applicant, start with 'Interview Feedback: 지원자님은' and express it in line text. Avoid using unprofessional or casual language and frame your advice using professional terminology appropriate to the interview situation. Please speak in Korean"},
                {"role": "user", "content": f"Please provide applicants with advice to help them pass the interview based on the following questions and answers. Please speak in Korean. \n질문: {row[0]}\n답변: {row[1]}"},
                {"role": "assistant", "content": row[3].replace("Interview Feedback: ", "면접 피드백: ")}
            ]
        }
        results.append(new_entry)

# 결과를 새로운 JSONL 파일로 저장
with open('/content/drive/MyDrive/Colab Notebooks/finalproject/questions_answers_final_feedback_100.jsonl', 'w') as outfile:
    for entry in results:
        json.dump(entry, outfile, ensure_ascii=False)
        outfile.write('\n')  # 각 JSON 객체 후에 줄바꿈 추가
```
